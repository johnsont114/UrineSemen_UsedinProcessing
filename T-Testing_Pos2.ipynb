```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
# Read the CSV file
df = pd.read_csv('Imputed_PMM_SemenAverageAreasProcessed.csv')
```

```python
df.head()
```

```python
dfH = df.loc[:,'Cornell_Pos_S-01_LMT_PS_PMM':'Cornell_Pos_S-50_LMH_LCT_PS_PMM']
#dfH = np.log10(dfH + 1)
```

```python
dfH.head()
```

```python
# Fetch the first row
first_row = dfH.iloc[0,:]

# Create a new Series to store the p-values
p_values = pd.Series(index=dfH.index)
logp_values = pd.Series(index=dfH.index)
r_coeff = pd.Series(index=dfH.index)

# Iterate over all subsequent rows
for i in range(1, len(df)):
    t_stat, p_value = ttest_ind(first_row, dfH.iloc[i, :])
    logp_values[i] = np.abs(np.log10(p_value))
    r_value = np.corrcoef(first_row, dfH.iloc[i, :])[0, 1]
    p_values[i] = p_value
    r_coeff[i] = r_value
    
# Add the p-values as a new column to the DataFrame
dfH['neg_logp_value'] = logp_values
dfH['p_value'] = p_values
dfH['r_coeff'] = r_coeff
```

```python
# Save the updated DataFrame back to the CSV file
dfH.to_csv('BC_Average_AllNeg_PMM_withP+r.csv')
```
