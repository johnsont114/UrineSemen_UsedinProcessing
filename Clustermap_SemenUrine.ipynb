```python
dfH = df.loc[:,'Cornell_Pos_S-01_LMT_PS_PMM_R1':'Cornell_Pos_S-50_LMH_LCT_PS_PMM_R1']
dfH = np.log10(dfH + 1)
#dfHB = dfH.loc[:, dfH.columns.str.contains('B')]
#dfHQ = dfH.loc[:, dfH.columns.str.contains('Q')]
dfHF = dfH.loc[:, dfH.columns.str.contains('O')]
#dfHR = dfH.loc[:, dfH.columns.str.contains('PM')]
```

```python
col_names = dfH.columns.values
col_namesDF = pd.DataFrame(col_names, columns=['columns'])
```

```python
col_namesDF['species'] = np.where(col_namesDF['columns'].str.contains('O'), 'dodgerblue', 'white')
col_namesDF = col_namesDF.set_index('columns')
col_namesDF

#below for more than two batches
#col_namesDF['species'] = np.where(col_namesDF['columns'], np.where(col_namesDF['columns'].str.contains('O'), 'dodgerblue', 'white'))
```

```python
sns.set(font_scale=1)
g = sns.clustermap(dfH, cmap='Purples', col_colors=col_namesDF, 
                   colors_ratio=(0.025, 0.025), standard_scale=0, 
                   dendrogram_ratio=0.08, cbar_pos=(-0.07, .45, .03, .2))
g.savefig('BC_Clustermap_Areas_30000_Full_SemenOnly_reruns_Pos2_Average_AllPeaks.png', dpi=300)
```

```python
clustermap_data = dfH.iloc[g.dendrogram_row.reordered_ind, g.dendrogram_col.reordered_ind]
```

```python
clustermap_data.to_csv('BC_CombinedReordered_Pos2Average_Clustermapdata.csv')
```
