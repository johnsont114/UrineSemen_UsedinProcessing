```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mass+Intensity.csv')
```

```python
data.head()
```

```python
for i in range (0, 187, 2):
    x = data.iloc[i].values[1:]  # This gets the first row
    y = data.iloc[i+1].values[1:]  # This gets the second row

    # Create scatter plot
    plt.scatter(x, y, label='Data Points', color='dimgray', s=1)
    plt.xlim(50,400)
    # Add vertical lines from the x-axis to each data point
    for xi, yi in zip(x, y):
        plt.vlines(xi, 0, yi, colors='black', linestyles='solid', alpha=0.7)

    # Add labels and title
    plt.xlabel('m/z')
    plt.ylabel('Intensity')
    plt.title(data.iloc[i].values[0])
    plt.xlim(50,500)
    #plt.ylim(0,max)
    #plt.legend()

    # Save the plot
    plt.savefig(f"{data.iloc[i].values[0]+str(i)}.png", dpi=300)

    # Clear the current figure's content for the next plot
    plt.clf()
```
