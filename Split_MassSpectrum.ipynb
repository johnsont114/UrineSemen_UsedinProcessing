```python
import pandas as pd
import csv
```

```python
# Load Excel file
df = pd.read_excel('S6_Semen_NewComps.xlsx')
```

```python
df.head()
```

```python
new_df = pd.DataFrame()
```

```python
for i, row in enumerate(df['MS/MS spectrum']):
    temp_data = []
    for item in row.split():
        num1, num2 = item.split(':')
        temp_data.append([num1, num2])

    temp_df = pd.DataFrame(temp_data, columns=[f'Mass_{i+1}', f'Intensity_{i+1}'])

    if new_df.empty:
        new_df = temp_df
    else:
        new_df = pd.concat([new_df.reset_index(drop=True), temp_df.reset_index(drop=True)], axis=1)
```

```python
new_df.to_csv('Mass+Intensity.csv', index=False)
```
